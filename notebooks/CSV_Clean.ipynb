# ALL OF THIS IS TO GET A SMALLER CSV TO WORK WITH!

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
datapath = 'C:\\Users\\Nick\\LendingClub Project\\New\\Data'
```

```python
df = pd.read_csv(datapath + '\\loan.csv', low_memory=False)
df.shape
```

```python
df.head()
```

## ONLY WANT LOANS THAT ARE FINALIZED (NO CURRENT LOANS)

```python
df['loan_status'].value_counts()
```

```python
loans_usable = ['Fully Paid', 'Charged Off', 'Default']
df1 = df[df['loan_status'].isin(loans_usable)]
df1.shape
```

```python
df1['loan_status'].value_counts()
```

## MANY COLUMNS WITH ~HALF NULL VALUES, DROP TO INCREASE PERFORMANCE

```python
(df1.isnull().sum() > 500000).value_counts()
```

## CREATE NEW DF WITH COLUMNS AND WHETHER THEY ARE MOSTLY NULL TO CREATE A LIST TO DROP

```python
clean = pd.DataFrame({'Name' : (df1.isnull().sum() > 500000).index.tolist(), 'Null' : (df1.isnull().sum() > 500000).tolist()})
clean.head()
```

## NEW DF TO EXPORT WITH CLEANED COLUMNS

```python
drop_null_cols = clean[clean['Null'] == True]['Name'].tolist()
x = df1.drop(drop_null_cols, axis=1)
```

```python
x.head()
```

```python
# Create new CSV to work with
```

```python
x.to_csv(datapath + '\\clean_loan.csv', index=False)
```
